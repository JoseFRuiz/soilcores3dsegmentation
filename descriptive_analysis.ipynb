# Descriptive analysis

```python
import os
import pandas as pd
```

```python
thrs = ['l0_u100', 'l0_u80', 'l0_u60', 'l0_u40', 'l0_u20', 'l20_u100', 'l40_u100', 'l60_u100', 'l80_u100']

dic_files_pearson = {}
dic_files_spearman = {}

for thr in thrs:
    dic_files_pearson[f"UNETR_v{thr}"] = os.path.join("outputs", "unetr_dataset_2adamw_100k_num_heads_2_vertical_png", f"correlation_pearson_{thr}_vertical.csv")
    dic_files_spearman[f"UNETR_v{thr}"] = os.path.join("outputs", "unetr_dataset_2adamw_100k_num_heads_2_vertical_png", f"correlation_spearman_{thr}_vertical.csv")
    dic_files_pearson[f"UNETR_h{thr}"] = os.path.join("outputs", "unetr_dataset_2adamw_100k_num_heads_2_horizontal_png", f"correlation_pearson_{thr}_horizontal.csv")
    dic_files_spearman[f"UNETR_h{thr}"] = os.path.join("outputs", "unetr_dataset_2adamw_100k_num_heads_2_horizontal_png", f"correlation_spearman_{thr}_horizontal.csv")
```

```python
corrs_pearson = {}
for key, value in dic_files_pearson.items():
    df = pd.read_csv(value)
    corrs_pearson[key] = df['Root Length Diameter Range 2 (px)'].loc[df["Unnamed: 0"]=='Sum of .L.>0.0000000'].values[0]
```
