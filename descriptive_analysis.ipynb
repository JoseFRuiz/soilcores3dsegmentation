```python
def plot_dictionary_barplot(data_dict, 
                          title="Bar Plot from Dictionary",
                          xlabel="Categories", 
                          ylabel="Values",
                          figsize=(10, 6),
                          color='skyblue',
                          edgecolor='black',
                          alpha=0.7,
                          rotation=45,
                          save_path=None,
                          show_plot=True,
                          sort_by='values',
                          ascending=True):
    """
    Plot a dictionary as a bar plot with optional sorting
    """
    import matplotlib.pyplot as plt
    import numpy as np
    
    # Sort the dictionary based on the sort_by parameter
    if sort_by == 'values':
        sorted_items = sorted(data_dict.items(), key=lambda x: x[1], reverse=not ascending)
    elif sort_by == 'keys':
        sorted_items = sorted(data_dict.items(), key=lambda x: x[0], reverse=not ascending)
    else:
        sorted_items = list(data_dict.items())
    
    # Extract keys and values from sorted items
    categories = [item[0] for item in sorted_items]
    values = [item[1] for item in sorted_items]
    
    # Create the plot
    plt.figure(figsize=figsize)
    bars = plt.bar(categories, values, 
                   color=color, 
                   edgecolor=edgecolor, 
                   alpha=alpha)
    
    # Customize the plot
    plt.title(title, fontsize=14, fontweight='bold')
    plt.xlabel(xlabel, fontsize=12)
    plt.ylabel(ylabel, fontsize=12)
    plt.xticks(rotation=rotation, ha='right')
    
    # Add value labels on top of bars
    for bar, value in zip(bars, values):
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2., height + height*0.01,
                f'{value:.2f}' if isinstance(value, (int, float)) else str(value),
                ha='center', va='bottom', fontsize=10)
    
    # Adjust layout to prevent label cutoff
    plt.tight_layout()
    
    # Save plot if path provided
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f"Plot saved to: {save_path}")
    
    # Show plot
    if show_plot:
        plt.show()
```

# Scatter plots

```python
def plot_scatter_pred_vs_gt(output_path, colname, gt_path, title, save_path):
    import matplotlib.pyplot as plt
    import pandas as pd
    import os
    import numpy as np
    from sklearn.linear_model import LinearRegression

    df_gt = pd.read_csv(gt_path)
    df_output = pd.read_csv(output_path)
    # Filter df_output to keep only rows where soil_core is in df_gt['corename']
    df_output_filtered = df_output[df_output['soil_core'].isin(df_gt['corename'])]

    x = df_gt['Sum of .L.>0.0000000'].values
    y = df_output_filtered[colname].values

    plt.figure(figsize=(12, 8))
    plt.scatter(x, y, alpha=0.7, s=100, label='Data points')
    plt.grid(True)
    plt.xlabel('Ground Truth', fontsize=24)
    plt.ylabel('Predicted', fontsize=24)
    plt.title(title, fontsize=24)

    # Fit a linear regression line
    x_reshaped = x.reshape(-1, 1)
    reg = LinearRegression().fit(x_reshaped, y)
    
    # Calculate R score
    from sklearn.metrics import r2_score
    y_pred = reg.predict(x_reshaped)
    
    # Plot the fitted line using the actual x data range
    x_min, x_max = min(x), max(x)
    x_line = np.linspace(x_min, x_max, 100)
    y_line = reg.predict(x_line.reshape(-1, 1))
    plt.plot(x_line, y_line, 'r--', linewidth=3, 
             label=f'Fitted line')
    
    plt.legend(fontsize=18)
    
    # Increase tick label sizes
    plt.xticks(fontsize=16)
    plt.yticks(fontsize=16)
    
    if save_path != None:
        plt.savefig(save_path)
    plt.show()
```

```python
title = "UNETR"
output_path = "outputs/dataset_2adamw_100k_num_heads_2_vertical_l20_u100_png/soil_cores_summary_vertical.csv"
gt_path = os.path.join('gt','CoresGT.csv')
colname = "Root Length Diameter Range 2 (px)"
save_path = f"outputs/{title}.png"

plot_scatter_pred_vs_gt(output_path, colname, gt_path, title, save_path)

title = "UNET"
output_path = "outputs/unet_unet_dataset_2_default_horizontal_l40_u100_png/soil_cores_summary_horizontal.csv"
gt_path = os.path.join('gt','CoresGT.csv')
colname = "Root Length Diameter Range 2 (px)"
save_path = f"outputs/{title}.png"

plot_scatter_pred_vs_gt(output_path, colname, gt_path, title, save_path)

title = "DynUNet"
output_path = "outputs/dynunet_dataset_2_100k_vertical_l20_u100_png/soil_cores_summary_vertical.csv"
gt_path = os.path.join('gt','CoresGT.csv')
colname = "Root Length Diameter Range 4 (px)"
save_path = f"outputs/{title}.png"

plot_scatter_pred_vs_gt(output_path, colname, gt_path, title, save_path)

title = "SegResNet"
output_path = "outputs/segresnet_segresnet_dataset_2_default_vertical_l20_u100_png/soil_cores_summary_vertical.csv"
gt_path = os.path.join('gt','CoresGT.csv')
colname = "Root Length Diameter Range 2 (px)"
save_path = f"outputs/{title}.png"

plot_scatter_pred_vs_gt(output_path, colname, gt_path, title, save_path)
```
